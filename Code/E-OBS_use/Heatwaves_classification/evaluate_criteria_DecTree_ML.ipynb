```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
import joblib
```

```python
df = pd.read_csv("/home/theom/Bureau/Ubuntu_SSD/PFE/Code/Van_der_Wiel_graphs/characteristics_htws.csv")
df
```

# Decision Tree Classification

```python
X = df.drop(columns=['extreme_bool','Impact_fct','Total_Deaths'])

y = df['extreme_bool']
model = DecisionTreeClassifier(max_depth=np.shape(X)[1], min_samples_leaf=1)

#all heatwaves

model.fit(X,y)

joblib.dump(model, '/home/theom/Bureau/Ubuntu_SSD/PFE/Code/Van_der_Wiel_graphs/Model/DTClass_all_characs.joblib')


tree.export_graphviz(model, out_file = '/home/theom/Bureau/Ubuntu_SSD/PFE/Code/Van_der_Wiel_graphs/Model/DTClass_all_characs.dot',
                    feature_names = X.columns, class_names=(y.unique()).astype(str),
                    label = 'all',rounded = True, filled=True)
```

```python
score = [0]*10000
for i in range(10000):
    X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
    model.fit(X_train,y_train)
    predictions = model.predict(X_test)
    score[i] = accuracy_score(y_test,predictions)
    #print(np.count_nonzero(y_test),np.count_nonzero(predictions))
print(np.max(score))
print(np.min(score))
print(np.mean(score))
print(np.std(score))
```
