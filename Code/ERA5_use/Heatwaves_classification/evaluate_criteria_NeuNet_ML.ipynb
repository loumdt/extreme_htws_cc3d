```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import tree
import matplotlib.pyplot as plt
import joblib
import random
from sklearn.base import BaseEstimator
```

```python
df = pd.read_csv("/home/theom/Bureau/Ubuntu_SSD/PFE/Code/Van_der_Wiel_graphs/characteristics_htws.csv")
df
```

```python
extreme_id = [123, 126, 132, 144, 146, 159, 161, 164, 171, 182, 194, 199, 201, 204, 209, 217, 237, 261, 270, 291, 295, 297, 305]
new_extreme_id = extreme_id
not_extreme_id = list(range(301))
for i in range(len(extreme_id)) : 
    new_extreme_id[i] = np.where(df.iloc[:,0]==extreme_id[i])[0][0]
    not_extreme_id.remove(new_extreme_id[i])
#print(new_extreme_id)


def split_train_test(X,y,extreme_size) :
    idx_list=list(range(301))
    sample1 = random.sample(new_extreme_id, extreme_size)
    sample2 = random.sample(not_extreme_id, 4*extreme_size)
    sample = sample1 + sample2
    for i in sample :
        idx_list.remove(i)
        
    X_train = X.iloc[sample,:]
    y_train = y.iloc[sample]
    X_test = X.iloc[idx_list,:]
    y_test = y.iloc[idx_list]

    return(X_train, X_test, y_train, y_test)

#X_train, X_test, y_train, y_test = split_train_test(X,y,15)
#print(X_train, y_train,)
```

```python
X = df.drop(columns=['extreme_bool','Impact_fct'])
y = df['extreme_bool']
model = MLPClassifier(hidden_layer_sizes=(1,),activation='logistic')
model.solver = 'lbfgs'
model.max_iter = 1e4

#X_train, X_test, y_train, y_test = split_train_test(X,y,15)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

model.fit(X_train,y_train)


predictions = model.predict(X_test)

print(y_test)

print(predictions)


score = accuracy_score(y_test,predictions)

print(np.max(predictions))

model.get_params(deep=True)
```
