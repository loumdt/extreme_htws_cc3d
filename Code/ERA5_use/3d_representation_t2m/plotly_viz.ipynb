```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np
import numpy.ma as ma
import netCDF4 as nc
import os
```

```python
lat = np.arange(10,20,1)
lon = np.arange(30,40,1)
time = np.arange(0, 3, 1)
lat=[0,1,2]
lon=[0,1,2]
time_mesh, lat_mesh, lon_mesh = np.meshgrid(time, lat, lon, indexing='ij')

temp = -9999*np.ones(time_mesh.shape)
temp[0,5,5]=10
temp[0,5,6]=10
temp[0,6,5]=10
temp[0,6,6]=10
temp[0,6,7]=10
temp[1,5,5]=10
temp[1,5,6]=10
temp[1,6,5]=10
temp[1,6,6]=10
temp[2,5,5]=10
temp[2,5,6]=10
temp[2,6,5]=10
temp[2,6,6]=10
temp[2,6,7]=10

fig = go.Figure(data=[go.Volume(x=lon_mesh, y=lat_mesh, z=time_mesh, value=lat_mesh, opacity=0.7, isomin=-15, isomax=15)])
#fig = go.Figure(data=None)
fig.update_layout(scene=dict(xaxis_title='Longitude', yaxis_title='Latitude', zaxis_title='Time'))
fig.show()
```
